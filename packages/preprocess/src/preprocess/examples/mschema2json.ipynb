```python
import re
import json

with open(
    "E:/code/AI/agentic-AI/SQL-QA/data/GSV/M-Schema/GSV-mysql-mschema.txt",
    "r",
    encoding="utf-8",
) as f:
    lines = f.readlines()
table_name = ""
tables = []
foreign_keys = []
for i, line in enumerate(lines[:]):
    line = line.strip()
    if "Foreign keys" in line:
        foreign_keys = list(l.strip() for l in lines[i + 1 :])
        break
    if re.search(r"^# Table: ", line):
        table_name = line.split("# Table: ")[1].strip()
        # print(table_name)

        table = dict()
        table["name"] = table_name
        tables.append(table)
    elif re.match(r"^\((.*)\)[^)]*", line):
        # print(line)
        matches = re.search(r"^\((.*?)\)[^)]*$", line)
        # print(matches.group(1))
        column = matches.group(1)
        col_name, desc = column.split(", ", 1)
        if "Examples: " in desc:
            desc, example = desc.split("Examples: ", 1)
            example = example.strip()
        else:
            example = None
        col_name, col_type = col_name.split(":", 1)
        table = tables[-1]
        if "columns" not in table:
            table["columns"] = []
        table["columns"].append(
            {
                "name": col_name.strip(),
                "description": desc.strip(),
                "type": col_type.strip(),
                "example": example,
            }
        )
print(json.dumps(tables, indent=4, ensure_ascii=False))
```

```python
json.dump(
    {"name": "GSV", "tables": tables, "foreign_keys": foreign_keys},
    open(
        "E:/code/AI/agentic-AI/SQL-QA/data/GSV/M-Schema/m-schema.json",
        "w",
        encoding="utf-8",
    ),
    ensure_ascii=False,
    indent=4,
)
```
